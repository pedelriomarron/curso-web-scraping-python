# Retener web con libreria socket

```python
import socket

mysock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
# Nos conectamos al puerto 80 de un servidor
mysock.connect(('www.w3.org', 80))

# Codificamos la peticion por get usando http
cmd='GET https://www.w3.org/TR/PNG/iso_8859-1.txt HTTP/1.0\r\n\r\n'.encode()
# enviamos la peticion
mysock.send(cmd)

# empezamos un bucle infinito
while True:
    # vamos sacando al tama√±o indicado los datos de la peticion
    data = mysock.recv(512)
    if len(data)<1:
        # cuando los datos sean menores de 0 acabamos bucle
        break
    print(data.decode())

# cerramos la conexion
mysock.close()
```

# Retener web con libreria urllib

```python
import urllib.request

f = urllib.request.urlopen('https://www.w3.org/TR/PNG/iso_8859-1.txt')

for line in f:
    # strip quitamos espacios en blanco
    print(line.decode().strip())
```
