```python
import urllib
from bs4 import BeautifulSoup

html = urllib.request.urlopen('https://openwebinars.net')
## convertimos codigo  a  objeto de beutifulSoup
soup = BeautifulSoup(html)
# Busca las etiquitas a
tags = soup('a')

for tag in tags:
    print('TAG: ',tag)
    print('URL: ',tag.get('href'))
    print('CONTENIDO: ',tag.contents)
    print('ATRIBUTOS: ',tag.attrs)
    print('\n')





```
