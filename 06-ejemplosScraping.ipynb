##  1 - Extracion del texto de un discurso

```python
# Instalacion de la libreria pandas
import sys
!{sys.executable} -m pip install pandas
```

```python
import urllib
from bs4 import BeautifulSoup
url = 'https://elpais.com/internacional/2018/01/31/actualidad/1517387619_036241.html'
html = urllib.request.urlopen(url)
soup2 = BeautifulSoup(html)

tags = soup2('p')
discurso = ""

for tag in tags:
    # Solo pasaria el corte si el tag p no tiene atributos
    if (len(tag.attrs))==0:
        a = tag.contents[0]
        # si es asi lo sumo al texto
        discurso += a

contadores = dict()
# borro signos
discurso = discurso.replace(',','').replace('.','').replace('?','').lower()
# meto cada palabra en uns poscicion de un array
palabras = discurso.split()

for palabra in palabras:
    if len(palabra)>3:
        # si la palabra exite suma 1
        contadores[palabra] = contadores.get(palabra,0) + 1
        
#Convertir diccionario en tabla
import pandas
pandas.DataFrame(list(contadores.items()), columns=['palabra','contador']).sort_values('contador',ascending=False)
```

##  2 - Extraer valor de un fondo de inversion

```python
participaciones = 50

url = 'http://www.morningstarfunds.ie/ie/funds/snapshot/snapshot.aspx?id=F00000PJME'

html = urllib.request.urlopen(url)
soup = BeautifulSoup(html)

tags = soup.find_all('td',class_="line text")
valor =  tags[0].contents

valor = float(valor[0].replace('EUR\xa0',''))

total = valor * participaciones

total
```

##  3 - Web scraping Baloncesto

```python
url = 'http://www.seleccionbaloncesto.es/inicio.aspx'

html = urllib.request.urlopen(url)
soup = BeautifulSoup(html)

#tags = soup.find_all('a',{"class":["", ""]})
tags = soup.findAll("td", id=lambda x: x and x.startswith('_ctl1_plantillaListView_') and x.endswith('_nombreTd'))

players = list()

for tag in tags:
    players.append(tag.contents[0].replace(" ","").replace("\r\n","").replace(",",""))

players




```
